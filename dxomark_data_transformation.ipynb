```python
# drop all null values
df_new = df.dropna(how="any")
df_new.head(3)
```

```python
# Creating column for chipset varient

df_new["Chip variant"] = df_new["Chipset"].apply(lambda x: str(x).split()[0])

df_new.shape
```

```python
# Making chipset varient same accross column

df_new['Chip variant'] = df_new['Chip variant'].apply(
    lambda x: "MediaTek" if "mediatek" in x.lower() or "mediaTek" in x else
              "Exynos"    if "samsung" in x.lower() else
              "HarmonyOS" if "huawei" in x.lower() else
              x
)
```

```python
# Clean MP and Aperture columns â†’ Convert to float

def clean_mp(value):
    if pd.isna(value):
        return None
    value = str(value).strip()
    value = re.sub("[^0-9.]", "", value)
    return float(value) if value.replace(".","",1).isdigit() else None

def clean_aperture(value):
    if pd.isna(value):
        return None
    value = str(value).lower().strip()
    value = value.replace("f/", "")
    value = value.replace(",", ".")
    value = re.sub("[^0-9.]", "", value)
    return float(value) if value.replace(".","",1).isdigit() else None


df_new["Camera MP"] = df_new["Camera MP"].apply(clean_mp)
df_new["Aperture"] = df_new["Aperture"].apply(clean_aperture)


print(df_new["Camera MP"].head())
print(df_new["Aperture"].head())
print(df_new.isnull().sum())
```

```python
#Adding missing MP values

rank_mp_map = {
    1: 12.5,
    22: 50,
    42: 48,
    61: 50,
    73: 50,
    94: 50,
    109: 50,
    112: 12.2,
    234: 50
}

df_new["Camera MP"] = df_new.apply(
    lambda row: int(round(rank_mp_map[row["Rank"]])) if row["Rank"] in rank_mp_map else row["Camera MP"],
    axis=1
)
```

```python
# Creating a new columns
# --> Image Processing Score (([Exposure] + [Color] + [Texture] + [Noise] + [Artifacts]) / 5)
# --> Balance Score ( ([Photo Score] + [Video Score] + [Bokeh Score]) / 3 )


df_new["Image Processing Score"] = (
    df_new[["Exposure", "Color", "Texture", "Noise", "Artifacts"]].mean(axis=1).round(1)
)


df_new["Balance Score"] = (
    df_new[["Photo Score", "Video Score", "Bokeh Score"]].mean(axis=1).round(1)
)

```

```python
df_new.to_csv("Final_dxoMark_scores.csv", index=False)
```
